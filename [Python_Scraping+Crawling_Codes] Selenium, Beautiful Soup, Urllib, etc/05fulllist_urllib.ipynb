```python
from urllib.request import urlopen #내장이라 별도 패키지 설치 불요!

url='http://www.hanbit.co.kr/store/books/full_book_list.html'
docs=urlopen(url)

encode=docs.info().get_content_charset()
text=docs.read().decode(encode)

with open('data/03fulllist.html','w',encoding=encode) as out: 
    out.write(text)
#웹의 페이지소스를 별도 파일에 몰아 저장하고!
with open('data/03fulllist.html','r',encoding=encode) as f:
    html=f.read()

import sys
import re

css1=r'<td class="left"><a.*?</td>'
css2=r'<td class="left">[^<>].*?</td>'
css3=r'<td class="right">[^<>].*?</td>'

titles=[]
writers=[]
prices=[]

for part_html in re.findall(css1,html,re.DOTALL):
    title=re.sub(r'<.*?>','',part_html)
    title=re.sub(r'&#40;','(',title) #괄호 표현 바꿔주기
    title=re.sub(r'&#41;',')',title)
    #print(title)
    titles.append(title)
print()
for part_html in re.findall(css2,html,re.DOTALL):
    writer=re.sub(r'<.*?>','',part_html)
    #print(writer)
    writer=writer.strip()
    writers.append(writer)
print()
for part_html in re.findall(css3,html,re.DOTALL):
    price=re.sub(r'<.*?>','',part_html)
    #print(price)
    price=price.strip()
    prices.append(price)
    
for i in range(len(titles)):
    print(titles[i],writers[i],prices[i])
    
with open('data/03fulllist.csv','w',encoding=encode) as out:
    fmt='%s,%s,%s\n' #줄바꾸기 유의!
    for i in range(len(titles)):
        out.write(fmt %(titles[i],writers[i].strip(),prices[i])) #이름 양옆 빈칸 없애기
```
