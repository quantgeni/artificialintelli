```python
#yes.24.co.kr>베스트
#도서제목저자정가 추출 //// #파일저장은 04yes24.html
#http://www.yes24.com/24/category/bestseller

from urllib.request import urlopen #내장이라 별도 패키지 설치 불요!

url='http://www.yes24.com/24/category/bestseller'
docs=urlopen(url)

encode=docs.info().get_content_charset()
text=docs.read().decode(encode)

print(encode)

with open('data/04yes24.html','w',encoding=encode) as out: 
    out.write(text)
#웹의 페이지소스를 별도 파일에 몰아 저장하고!
with open('data/04yes24.html','r',encoding=encode) as f:
    html=f.read()

import sys
import re

#<a href="http://www.qnk">나태주</a> 저 |

css1=r'\[도서.*?</a>' #[ <- 이걸 escape 시켜주기 위해 \ 붙여주는 것에 유의!!

#css1=r'\[도서\] <a href=.*?</a>'
css2=r'<p class="aupu"><a.*?\|'
css3=r'<p class="price"><strong>.*?</strong.*?>'

titles=[]
writers=[]
prices=[]

for part_html in re.findall(css1,html,re.DOTALL):
    title=re.sub(r'<.*?>','',part_html)
    title=re.sub(r'&#40;','(',title) #괄호 표현 바꿔주기
    title=re.sub(r'&#41;',')',title)
    title=re.sub(r'\[도서\]','',title)
    title=title.strip() #<- 유의!!! 공백 없애줌
    #print(title)
    titles.append(title)
print()
for part_html in re.findall(css2,html,re.DOTALL):
    writer=re.sub(r'<.*?>','',part_html)
    writer=re.sub(r'\|','',writer)
    #riter=re.sub(r',','',writer) <- 저자 여러명인 경우 때문인데, 굳이 이렇게 안 하고 밑에서 아예 큰 따옴표로 분리함
    writer=writer.strip()
    #print(writer)
    writers.append(writer)
print()
for part_html in re.findall(css3,html,re.DOTALL):
    price=re.sub(r'<.*?>','',part_html)
    price=re.sub(r',','',price)
    #print(price)
    price=price.strip()
    prices.append(price)
    
for i in range(len(titles)):
    print(titles[i],writers[i],prices[i])
    
with open('data/04yes24.csv','w',encoding='utf-8') as out: #encoding=encode로 안 하고 'utf-8'로 찍어줌
    fmt='%s,"%s",%s\n' #줄바꾸기 유의!
    for i in range(len(writers)):
        out.write(fmt %(titles[i],writers[i],prices[i]))
```
