```python
#온라인 도서 쇼핑몰의 도서 정보를 스크래핑. 먼저 도서정보가 있는 사이트의 url을 알아냄!!
#hanbit.co.kr -> store -> bestseller -> 도서 목록 제목 추출
#http://www.hanbit.co.kr/store/books/bestseller_list.html
#02bestseller.html
from urllib.request import urlopen #내장이라 별도 패키지 설치 불요!

url='http://www.hanbit.co.kr/store/books/bestseller_list.html'
docs=urlopen(url)
#스크래핑 할 웹페이지의 url을 urlopen 함수의 매개변수로 전달함
#print(docs) #html 소스가 객체 형태로 저장되어 있음
#text=docs.read()
#print(text) #객체 형태로 저장된 내용을 read함수를 이용해서 텍스트로 변환 후 출력
#하지만 인코딩 처리 없이 변환했기 때문에 한글은 제대로 보이지 않음

encode=docs.info().get_content_charset()
#http 헤더를 기반으로 스크래핑한 문서의 인코딩 방식을 알아냄
text=docs.read().decode(encode)

with open('data/02bestseller.html','w',encoding=encode) as out: 
    out.write(text)
#웹의 페이지소스를 별도 파일에 몰아 저장하고!
with open('data/02bestseller.html','r',encoding='utf-8') as f:
    html=f.read()
#위에 몰아 저정한 내용들을 한줄씩 읽으면서 html에 다 쳐박아 놓음    
#도서 제목 추출 위해 정규표현식 이용. 파이썬 정규표현식 패키지는?? => re

#from urllib.request import urlopen
import sys
import re

#스크래핑한 결과를 저장한 파일로부터 내용을 읽어들여 화면에 출력
css1=r'<p class="book_tit"><a.*?</p>'
css2=r'<p class="book_writer">.*?</p>' #작가 가져오기. 홈페이지 소스 보면 title과 다르게 a태그 없는 것을 알 수 있다

for part_html in re.findall(css1,html,re.DOTALL):
    title=re.sub(r'<.*?>','',part_html)
    print(title)
print()
for part_html in re.findall(css2,html,re.DOTALL):
    writer=re.sub(r'<.*?>','',part_html)
    print(writer)
#위에서 read 하면서 데이터 쳐박아넣은 html 상에서 htm만을 찾아서 sub으로 없애서 title에 담음
```
