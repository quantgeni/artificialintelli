```python
#사원클래스 정의
#사원클래스에는 이름,나이,성별,급여,입사일 등의 속성을 가짐.
#한편, 상사 클래스에는 이름,나이,성별,급여,입사일,직위,근무시간 등의 속성을 가짐

#Inheritance 없이
class Employee:
    def __init__(self,name,age,gender,salary,hdate):
        self.name=name
        self.age=age
        self.gender=gender
        self.salary=salary
        self.hdate=hdate
 

    def __str__(self):
        msg='%s,%s,%s,%s,%s' %(self.name,self.age,self.gender,self.salary,self.hdate)
        return msg

class Manager:
    def __init__(self,name,age,gender,salary,hdate,mgrid,wtime):
        self.name=name
        self.age=age
        self.gender=gender
        self.salary=salary
        self.hdate=hdate
        self.mgrid=mgrid
        self.wtime=wtime
    def __str__(self):
        msg='%s,%s,%s,%s,%s,%s,%s' %(self.name,self.age,self.gender,self.salary,self.hdate,self.mgrid,self.wtime)
        return msg
    
emp1=Employee('수지',25,'여',25000,'2019-03-03')
print(emp1)

mgr1=Manager('혜교',35,'여',25000,'2020-03-01',35,10)
print(mgr1)

#만일 사원/상사 클래스에 전화번호,이메일 속성 추가하려면? 둘 모두에 멤버변수 추가해야 함
#만일 둘 클래스에 성별 속성 제거한다면? 둘 모두 gender 멤버변수 제거해야 함
# => 결국 불편
```

```python
class Person:
    def __init__(self,name,age,gender,salary,hdate):
        self.name=name
        self.age=age
        self.gender=gender
        self.salary=salary
        self.hdate=hdate
 

    def __str__(self):
        msg='%s,%s,%s,%s,%s' %(self.name,self.age,self.gender,self.salary,self.hdate)
        return msg 

class Employee2():
    def __init__(self,name,age,gender,salary,hdate):
        self.name=name
        self.age=age
        self.gender=gender
        self.salary=salary
        self.hdate=hdate

    def __str__(self):
        msg='%s,%s,%s,%s,%s' %(self.name,self.age,self.gender,self.salary,self.hdate)
        return msg

class Employee2(Person): #Person 클래스를 상속받아 Employee2 클래스 정의
    pass

class Manager2(Person): #Person 클래스를 상속받아 Manager2 클래스 정의
    def __init__(self,name,age,gender,salary,hdate,mgrid,wtime):
        super().__init__(name,age,gender,salary,hdate) #super라는 지시자 통해 상위 클래스 생성자/메서드 참조 가능
                                                       #괄호 안에 self 없는 이유는 super() <- 이놈이 self 인자로 들어가서 개수가 맞음
        self.mgrid=mgrid
        self.wtime=wtime
    
    def __str__(self):
        msg=super().__str__()+'%s,%s' %(self.mgrid,self.wtime)
        return msg
   
    
psn=Person('수지',25,'여',25000,'2019-03-03')
print(psn)

mgr1=Manager2('혜교',35,'여',350000,'2010-05-05',35,10)
print(mgr1)
```

## 상속이 진행될수록 상위클래스 기능은 일반적이 되고 하위클래스 기능은 구체적이 됨
## 또한 상위 클래스 객체보다 하위 클래스 객체가 좀더 많은 정보 갖고 있음

```python
# 스타크래프트 유닛을 상속개념 이용하여 구현
# 테란 진영 유닛을 클래스로 작성
# 해병대 : 이름 체력 에너지 공격력 10 플라즈마소총
# 의무관 : 이름 체력 에너지 공격력1 치료
# 화염방사병 : 이름 체력 에너지 공격력20 화염방사기

class 해병대: #공격력 10짜리 유닛
    def __init__(self,life,energy):
        self.name=self.__class__.__name__
        self.life=life
        self.energy=energy
        self.attack=10
    def __str__(self):
        msg='이름:%s, 라이프:%s, 에너지:%s, 공격력:%s' %(self.name,self.life,self.energy,self.attack)
        return msg
    def attacks(self):
        print('삼엄한 경계 발동 중')
        
class 의무관: #공격력 1짜리 유닛
    def __init__(self,life,energy):
        self.name=self.__class__.__name__
        self.life=life
        self.energy=energy
        self.attack=1
    def __str__(self):
        msg='이름:%s, 라이프:%s, 에너지:%s, 공격력:%s' %(self.name,self.life,self.energy,self.attack)
        return msg
    def heals(self):
        print('힐링 담당')
        
class 화염방사병: #공격력 20짜리 유닛
    def __init__(self,life,energy):
        self.name=self.__class__.__name__
        self.life=life
        self.energy=energy
        self.attack=20
    def __str__(self):
        msg='이름:%s, 라이프:%s, 에너지:%s, 공격력:%s' %(self.name,self.life,self.energy,self.attack)
        return msg
    def attacks(self):
        print('ready to fire')
        
marine1=해병대(100,100)
print(marine1)
marine1.attacks()

medic1=의무관(200,200)
print(medic1)
medic1.heals()

fire1=화염방사병(70,70)
print(fire1)
fire1.attacks()
```

```python
#class+inheritance

class Unit:
    def __init__(self,life,energy):
        self.name=self.__class__.__name__
        self.life=life
        self.energy=energy
    def __str__(self):
        msg='%s,%s,%s'%(self.name,self.life,self.energy)
        return msg
    def attack(self):
        pass
class 해병대(Unit): #공격력 10짜리 유닛
    def __init__(self,life,energy):
        super().__init__(life,energy)
        #부모클래스에 정의된 생성자를 호출해서 변수 초기화
        self.attack=10
    def __str__(self):
        #method overriding => 상속관계 상위클래스에서 정외된 함수를 하위 클래스에 적합하도록 재작성하는 것을 의미
        msg=super().__str__()
        return '%s, %s' %(msg,self.attack)
    def attacks(self):
        print('삼엄한 경계 발동 중')

        
class 의무관(Unit): #공격력 1짜리 유닛
    def heals(self):
        print('힐링 담당')
        
class 화염방사병(Unit): #공격력 20짜리 유닛
    def attacks(self):
        print('ready to fire')
      
marine2=해병대(300,100)
print(marine2)
marine2.attacks()

medic2=의무관(500,50)
print(medic2)
medic2.heals()
```

```python
#성적 프로그램을 상속을 적용해서 작성
#중간고사성적:이름 국어 영어 수학
#기말고사성적: 이름 국어 영어 수학 과학 미술
#공통 요소: 총점 평균 학점
#성적처리 메서드 computeSungJuk

#중간고사 혜교 99 98 97
#기말고사 지현 65 84 90 50 55

#상위 클래스
class sungJuk:
    def __init__(self,name,kor,eng,mat):
        #사용자가 입력한 값으로 멤버변수 초기화
        self.name=name
        self.kor=kor
        self.eng=eng
        self.mat=mat
        #초기값으로 멤버변수 생성
        self.tot=0
        self.mean=0.0
        self.grd='가'
        
    def computeSungJuk(self):
        self.tot=self.kor+self.eng+self.mat
        self.mean=self.tot/3
        if self.mean>=90:self.grd='수'
        elif self.mean>=80:self.grd='우'
        elif self.mean>=70:self.grd='미'
        elif self.mean>=60:self.grd='양'
        elif self.mean>=50:self.grd='가'
        else:self.grd='공부해'
        
    def __str__(self):
        msg='%s,%d,%d,%d,%d,%.1f,%s' %(self.name,self.kor,self.eng,self.mat,self.tot,self.mean,self.grd)
        return msg
#==================================================================
#하위 클래스      
class MidSungJuk(sungJuk):
        pass
        
class FinalSungJuk(sungJuk):
        def __init__(self,name,kor,eng,mat,sci,art):
            super().__init__(name,kor,eng,mat)
            self.sci=sci
            self.art=art    
        def __str__(self):
            msg=super().__str__()
            msg='%s,%s,%s' %(msg,self.sci,self.art)
            return msg
        def computeSungJuk(self):
            super().computeSungJuk()
            self.tot=self.tot+self.sci+self.art
            self.mean=self.tot/5
            if self.mean>=90:self.grd='수'
            elif self.mean>=80:self.grd='우'
            elif self.mean>=70:self.grd='미'
            elif self.mean>=60:self.grd='양'
            elif self.mean>=50:self.grd='가'
            else:self.grd='공부해'          

msj=MidSungJuk('혜교',30,50,60)
msj.computeSungJuk()
print(msj)

fsj=FinalSungJuk('지현',79,78,75,53,55)
fsj.computeSungJuk()
print(fsj)
```

```python
from abc import *

class Phone(metaclass=ABCMeta):
    def __init__(self,maker,price,comm,os):
        self.maker=maker
        self.price=price
        self.comm=comm
        self.os=os
        
    def __str__(self):
        msg='%s,%s,%s,%s' %(self.maker,self.price,self.comm,self.os)
        return msg
    
    @abstractmethod
    def notworked(self):
        pass
    
class Phone4G(Phone):
    def __init__(self,maker,price,comm,os,dnsp,upsp):
        super().__init__(maker,price,comm,os)
        self.dnsp=dnsp
        self.upsp=upsp
        
    def __str__(self):
        msg=super().__str__()
        msg='%s,%s,%s' %(msg,self.dnsp,self.upsp)
        return msg

    def notworked(self):
        print('')
    
class Phone5G(Phone):
    def __init__(self,maker,price,comm,os,autocar,ai):
        super().__init__(maker,price,comm,os)
        self.autocar=autocar
        self.ai=ai
        
    def __str__(self):
        msg=super().__str__()
        msg='%s,%s,%s' %(msg,self.autocar,self.ai)
        return msg

    def notworked(self):
        print('')
    
ph1=Phone4G('Google',980000,'LTEAdv','Oreo','1Gbps','500Mbps')
ph2=Phone5G('LG',1300000,'NR2019','PIE','Audi240','Amazon')
print(ph1)
print(ph2)
```
