# readline/readlines vs. for ~ in f:

## readline/readlines 개선 버전
## 대용량 파일을 위 방식으로 읽는 경우 memoryError 유발될 수 있음
## f.readline()은 파일 모든 내용을 메모리 상 적재 후 한줄씩 읽어 처리!
## 반면, for ~ in f는 순환자(iterator) 형식.. 파일객체가 가리키는 대상으로부터 한줄씩 읽어옴
###  (즉, 후자는 메모리에 다 올려놓지 않고, 보조기억장치로부터 읽어옴...)

=> 즉, 양이 적으면 realine() 오키. 양이 많으면 for ~in f 형식 좋음. 그냥 for ~in f로 퉁쳐서 사용이 굿.

# iterator : 반복자, 순환자
# => 컬렉션에 저장된 데이터를 순차적으로 접근하면서 처리하는 데이터 형식. 데이터 유형 상관없이 동일한 방식으로
# 코드를 적용시켜 프로그래밍 할 수 있게 해줌 
### for ~ in f: 에서 f가 리스트건 문서건 상관없다 이것임

```python
# 성적 처리 프로그램. 3명 학생으로부터 이름 국어 영어 수학 입력받아 총점 평균 학점 계산 후 sungjuks.txt라는 파일에 저장

with open('sungjuks.txt','a',encoding='utf-8') as f:
    for pizza in range(3):
        name=input('이름은?')
        kor=int(input('국어성적?'))
        eng=int(input('영어성적?'))
        mat=int(input('수학성적?'))
        tot=kor+eng+mat
        mean=tot/3
        grd='천재'
        if mean>=90:grd='수'
        elif mean>=80:grd='우'
        elif mean>=70:grd='미'
        elif mean>=60:grd='양'
        else:grd='공부좀해'
        f.write('이름:%s,국어:%d,영어:%d,수학:%d,총점:%d,평균:%.1f,등급:%s\n' %(name,kor,eng,mat,tot,mean,grd))
```

```python
#sungjukdata.txt (3명 학생 데이터) 이용해서 총점 평균 학점 계산 후 sungjuks2.txt라는 파일에 저장
import re
i=0
datalist=[]

def computeSungjuk(k,e,m):
        tot=k+e+m
        mean=tot/3
        grd='가'
        if mean>=90:grd='수'
        elif mean>=80:grd='우'
        elif mean>=70:grd='미'
        elif mean>=60:grd='양'
        else:grd='공부좀해'
        return tot,mean,grd

with open('sungjukdata.txt','r',encoding='utf-8') as f:
    for data in f:
        data=re.sub(r'\n','',data) #중요!! 위에 re 임포트 해서 줄바꾸기 없앰!! 당초 파일에서 각 줄 마지막 단어에 \n 없애주기 위함임
        sj=data.split(',') #한줄씩 읽되 파일에서 읽은 데이터를 컴마로 구분해서 리스트화 시킴
        i+=1
        print(i)
        print(sj[0],sj[1],sj[2],sj[3])
        tot,mean,grd=computeSungjuk(int(sj[1]),int(sj[2]),int(sj[3])) #int화 시키는거랑 def된 함수의 return값들을 별도 변수에 담는거 유의!
        result='%s,%s,%s,%s,%d,%.1f,%s\n' %(sj[0],sj[1],sj[2],sj[3],tot,mean,grd)
        datalist.append(result)
print(datalist)
    
with open('sungjuks2.txt','w',encoding='utf-8') as f:
    for obj in datalist:
        f.write(obj)  #주목할만한 점은, 각 obj는 ' '로 되어 있지만 write 해보면 ' ' 없이 들어감. str을 기본값으로 놓기 때문인 것 같음
```

```python
#sungjukdata.txt (3명 학생 데이터) 이용해서 총점 평균 학점 계산 후 sungjuks2.txt라는 파일에 저장
#파일인코딩이 utf-8인 경우 문서 맨 처음에 \ufeff가 추가되어 있음. 
#이것이 거슬릴 경우 파일 인코딩을 utf-8-sig로 설정하면 해당 문자열이 안 보임!!

import re
i=0
datalist=[]

def computeSungjuk(k,e,m):
        tot=k+e+m
        mean=tot/3
        grd='가'
        if mean>=90:grd='수'
        elif mean>=80:grd='우'
        elif mean>=70:grd='미'
        elif mean>=60:grd='양'
        else:grd='공부좀해'
        return tot,mean,grd

with open('sungjukdata.txt','r',encoding='utf-8-sig') as f:
    for data in f:
        #data=re.sub(r'\n','',data) #중요!! 위에 re 임포트 해서 줄바꾸기 없앰!! 당초 파일에서 각 줄 마지막 단어에 \n 없애주기 위함임
        sj=data.strip().split(',') #한줄씩 읽되 파일에서 읽은 데이터를 컴마로 구분해서 리스트화 시킴
        #strip()을 중간에 삽입하여 좌우공백 제거하면 굳이 re 임포트 해서 \n를 ''로 교체하지 않아도 됨.
        #결국 문서에서 줄바뀜을 인지하는게 각 줄 마지막 글자 오른쪽이 공백이면 그렇다는건가... 그걸 없애니 \n로 인지하지 않음? 아무튼.
        i+=1
        print(i)
        print(sj[0],sj[1],sj[2],sj[3])
        tot,mean,grd=computeSungjuk(int(sj[1]),int(sj[2]),int(sj[3])) #int화 시키는거랑 def된 함수의 return값들을 별도 변수에 담는거 유의!
        result='%s,%s,%s,%s,%d,%.1f,%s\n' %(sj[0],sj[1],sj[2],sj[3],tot,mean,grd)
        datalist.append(result)
print(datalist)
    
with open('sungjuks2.txt','w',encoding='utf-8') as f:
    for obj in datalist:
        f.write(obj)  #주목할만한 점은, 각 obj는 ' '로 되어 있지만 write 해보면 ' ' 없이 들어감. str을 기본값으로 놓기 때문인 것 같음
```

```python
#iris.csv 파일을 이용해서 붓꽃의 너비/길이, 꽃받침의 너비/길이에 대해 
#총합,평균,최대,최소,중앙값,표준편차,표본분산을 계산하고
#descpStastic.txt라는 파일에 저장

import statistics
    
data1=[]
data2=[]
fdata1=[]
fdata2=[]


def calcul(d):
    tot=sum(d)
    mean=tot/len(d)
    maxnum=max(d)
    minnum=min(d)
    median=statistics.median(d)
    stdev=statistics.stdev(d)
    variance=statistics.variance(d)
    return tot,mean,maxnum,minnum,median,stdev,variance

with open('iris.csv','r',encoding='utf-8-sig') as f:
    f.readline()
    for eline in f:
        flow=eline.strip().split(',')
        c1=float(flow[1])/float(flow[0])
        c2=float(flow[3])/float(flow[2])
        data1.append(c1)
        data2.append(c2)
        tot,mean,maxnum,minnum,median,stdev,variance=calcul(data1)
        result1='%.1f,%.1f,%.1f,%.1f,%.1f,%.1f,%.1f'%(tot,mean,maxnum,minnum,median,stdev,variance)
        tot,mean,maxnum,minnum,median,stdev,variance=calcul(data2)
        result2='%.1f,%.1f,%.1f,%.1f,%.1f,%.1f,%.1f'%(tot,mean,maxnum,minnum,median,stdev,variance)
        fdata1.append(result1)
        fdata2.append(result2)

    
with open('descpStastic2.txt','a',encoding='utf-8-sig') as f:
    for each in fdata1:
        f.write(each)
    for each in fdata2:
        f.write(each)
```

```python
with open('iris.csv','r',encoding='utf-8') as f:
    f.readline() #첫째줄은 스킵
    #cnt=0 #데이터 행 수 위한 변수
    irisds=[]
    pls=[];pws=[];sls=[];sws=[];
    for rec in f:
        #print(rec)
        iris=rec.strip().split(',')
        cnt+=1
        pls.append(float(iris[0]))
        pws.append(float(iris[1]))
        sls.append(float(iris[2]))
        sws.append(float(iris[3]))

from statistics import mean,median,stdev,variance
print('총 데이터 수: ',cnt)
print('%.1f, %.1f, %.1f, %.1f' %(sum(sls),sum(sws),sum(pls),sum(pws)))
print('%.1f, %.1f, %.1f, %.1f' %(max(sls),max(sws),max(pls),max(pws)))
print('%.1f, %.1f, %.1f, %.1f' %(min(sls),min(sws),min(pls),min(pws)))
print('%.1f, %.1f, %.1f, %.1f' %(mean(sls),mean(sws),mean(pls),mean(pws)))
print('%.1f, %.1f, %.1f, %.1f' %(median(sls),median(sws),median(pls),median(pws)))
print('%.1f, %.1f, %.1f, %.1f' %(stdev(sls),stdev(sws),stdev(pls),stdev(pws)))
print('%.1f, %.1f, %.1f, %.1f' %(variance(sls),variance(sws),variance(pls),variance(pws)))      

irisds.append('DS,SepalLength,SepalWidth,PetalLength,PetalWidth\n')
irisds.append('count,%d개,%d개,%d개,%d개\n' %(cnt,cnt,cnt,cnt))
irisds.append('sum,%.1f,%.1f,%.1f,%.1f\n' %(sum(sls),sum(sws),sum(pls),sum(pws)))     
irisds.append('max,%.1f,%.1f,%.1f,%.1f\n' %(max(sls),max(sws),max(pls),max(pws)))
irisds.append('min,%.1f,%.1f,%.1f,%.1f\n' %(min(sls),min(sws),min(pls),min(pws)))
irisds.append('mean,%.1f,%.1f,%.1f,%.1f\n' %(mean(sls),mean(sws),mean(pls),mean(pws)))
irisds.append('median,%.1f,%.1f,%.1f,%.1f\n' %(median(sls),median(sws),median(pls),median(pws)))
irisds.append('stdev,%.1f,%.1f,%.1f,%.1f\n' %(stdev(sls),stdev(sws),stdev(pls),stdev(pws)))
irisds.append('variance,%.1f,%.1f,%.1f,%.1f\n' %(variance(sls),variance(sws),variance(pls),variance(pws)))   

print(irisds)

with open('descpStastic3.txt','w',encoding='utf-8') as f:
    for obj in irisds:
        f.write(obj)
```

```python
#키보드로 입력받은 내용을 숫자로 변환하는 코드를 예외처리하세요
#abc 입력 시 "숫자만 입력"
#[1,2,3] 입력 시 "스칼라형 숫자만 입력하세요"
def myconvert(i):
    out=0;
    try:
        out=int(i)
    except ValueError:
        print('숫자만 입력하세요!')
    except TypeError:
        print('단일 숫자형문자만 입력하세요!')
    return out

print(myconvert('123'))
print(myconvert(123))
print(myconvert('ABC'))
print(myconvert('[1,2,3]'))
```

```python
#checked exception vs. unchecked exception(문법상의문제가없어서예외체크는못하고실행해보아야만알수있는예)

try:
    import xyz
    print([1,2,3][3])
    print('a'+1)
    print(4/0)
except ModuleNotFoundError:
    print('해당모듈이 없어요')
except IndexError:
    print('요소의 인덱스 지정이 잘못되었어요!')
except TypeError:
    print('숫자와 문자는 더할 수 없어요!')
except ZeroDivisionError:
    print('0으로 나눌 수 없어요!')
```
