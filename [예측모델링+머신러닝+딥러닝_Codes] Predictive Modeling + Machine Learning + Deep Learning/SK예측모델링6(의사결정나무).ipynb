```
# 예제 데이터 생성 및 시각화

X, y = make_blobs(n_samples=300, centers=4, cluster_std=1.0, random_state=1907081430)
plt.scatter(X[:,0], X[:,1], c=y, cmap='rainbow', s=50) # 색상은 y변수에 따라 (c=y). cmap 굳이 안 써도 됨. // s는 점의 크기
plt.show()

print(y)
```

```
# 의사결정트리로 분류하기
# 이후, 결정영역 표시하기

dtc = DecisionTreeClassifier()
dtc.fit(X,y)

plot_decision_regions(X,y, clf=dtc)
plt.show()

print('훈련정확도', dtc.score(X, y)) # ㅋㅋ 1 (100%) 뜸
```

```
# 훈련/검증 데이터 분리
data = X
target = y

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=1907081500)

dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
pred = dtc.predict(X_test)
```

```
# 분석결과 평가
print('정확도', accuracy_score(pred, y_test))
print(confusion_matrix(pred, y_test))

plot_decision_regions(X_test, y_test, clf=dtc)
plt.show()
```
