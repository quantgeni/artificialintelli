```
a = pd.DataFrame([[1, np.nan, 3],
                [np.nan, 8, 10],
                [15, 20, 18]])
print('null 출력', a.isna())
print('----------------------------------------')
print('정상값 출력 1', a[a.notna()])
print('정상값 출력 2', a[~a.isna()]) #데이터프레임에 notna()를 이용해서 정상값을 출력하기는 어려움
print('----------------------------------------')
print('null 제거1', a.dropna())
print('null 제거2', a.dropna(axis=0)) 

# 데이터프레임에 dropna()를 사용하면
# 기본적으로 행을 기준으로 null이 삭제 (즉, axis=0을 지정한 것과 동일)
print('----------------------------------------')
print('null 제거2', a.dropna(axis=1)) # 열 기준 null 삭제
print('----------------------------------------')
# null을 다른 값으로 대체하기
print('null 대체1', a.fillna(0))
print('----------------------------------------')
print('null 대체2', a.fillna(method='ffill'))
print('null 대체2', a.fillna(method='bfill'))
# null 값 기준 앞/뒤 값으로 대체함
print('---------------------------------------- full version')
print(a)
print('---------------------------------------- ffill, 행 기준')
print('null 대체2', a.fillna(method='ffill', axis=0))
print('---------------------------------------- bfill, 행 기준')
print('null 대체2', a.fillna(method='bfill', axis=0))
print('---------------------------------------- ffill, 열 기준')
print('null 대체2', a.fillna(method='ffill', axis=1))
print('---------------------------------------- bfill, 열 기준')
print('null 대체2', a.fillna(method='bfill', axis=1))
print('---------------------------------------- 위에 다 귀찮으면 그냥 평균값으로 NA 대체하는 것도 무난함')
```

```
# pandas 기술 통계 다루기

name = ['Tom','James','Ricky','Vin','Steve','Smith','Jack','Lee','David','Gasper','Betina','Andres']
age = [25,26,25,23,30,29,23,34,40,30,51,46]
rating = [4.23,3.24,3.98,2.56,3.20,4.6,3.8,3.78,2.98,4.80,4.10,3.65]
data = {'name':name, 'age':age, 'rating':rating}

df = pd.DataFrame(data)
print(df)
print('---------------------------------------------')
# 데이터 행수 : count
print('전체 행수', df.count())
print('---------------------------------------------')
# 전체 합 : sum(축)
print('전체 합', df.sum()) # 열 기준
print('전체 합', df.sum(1)) # 행 기준

print('---------------------------------------------')
# 평균 : mean
print('평균', df.mean())
print('---------------------------------------------')
```

```
# 데이터프레임에서 순위rank 다루기
# 기본적으로 오름차순으로 출력
# 동일한 값들은 그것들의 평균값으로 지정
# 생각하기 귀찮으니 아래 rank()는 일단 제끼자

data = [-7, 2, 7, 4, 2, 0, 4]
df =  pd.Series(data)
print(df)
print('---------------------------------------------')
print(df.rank())
print('---------------------------------------------')
print(df.rank(method='first')) # 동일순위의 경우 데이터 순위가 기준
print('---------------------------------------------')
print(df.rank(method='max')) # => ascending 옵션 + method 옵션의 간략화 버전
print(df.rank(method='min')) # => ascending 옵션 + method 옵션의 간략화 버전
print('---------------------------------------------')
print(df.rank(ascending= False, method='max')) # 동일순위의 경우 큰 데이터 기준 => ascending 옵션 + method 옵션의 간략화 버전
```

```
# 데이터 정렬하기


# 인덱스로 정렬 : sort_index

unsorted = pd.DataFrame(np.random.randint(1,46, (5,2)),
                       index=[1,3,2,5,4],
                        columns=['나','가']
                       )
print(unsorted)
print('---------------------------------------------')
# 인덱스로 정렬하기
print(unsorted.sort_index()) # 아무것도 안 쓰면 default axis=0 행 인덱스 기준 정렬
print(unsorted.sort_index(axis=1)) # axis=1 열 인덱스 기준 정렬
print(unsorted.sort_index(ascending=False))
```

```
# 값으로 정렬 : sort_values
# 단, 정렬 시 기준열을 by 속성으로 지정해야 함

unsorted = pd.DataFrame(np.random.randint(1,46, (5,2)),
                       index=[1,3,2,5,4],
                        columns=['가','나']
                       )
print(unsorted)
print('---------------------------------------------')
# 인덱스로 정렬하기
print(unsorted.sort_values(by='가'))
print(unsorted.sort_values(by='나'))
print(unsorted.sort_values(by=['가','나']))
print(unsorted.sort_values(by=['가','나'],
                          ascending=False))
```

```
# 데이터프레임 그룹핑
# 특정 조건에 따라 데이터들을 묶어서 집계함수를 적용해서 통계정보를 출력
# 해당 그룹의 특성을 살펴볼 수 있음
# groupby(조건열).(집계함수)

city = ['서울', '서울', '서울', '부산', '부산', '부산', '인천', '인천']
year = [2015, 2010, 2005, 2015, 2010, 2005, 2015, 2010]
pop = [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203]
region = ['수도권', '수도권', '수도권', '경상권', '경상권', '경상권', '수도권', '수도권']

data = {'도시':city,'연도':year,'인구':pop,'지역':region}

df = pd.DataFrame(data, index=np.arange(1,9))
print(df)
print('----------------------------------------------------------------------')
# 도시 별로 그룹핑
groupby = df.groupby('도시')
print(groupby)
print('----------------------------------------------------------------------')
print(groupby.groups) # 그룹핑 된 결과 출력
print('----------------------------------------------------------------------')
for name, group in groupby:
  print(name)
  print(group)
print('----------------------------------------------------------------------')

# 하기 출력 결과보면 연도도 연산됨

print(groupby.agg(np.sum))
print('----------------------------------------------------------------------')
print(groupby.agg(np.mean))
print('----------------------------------------------------------------------')
print(groupby.agg(np.min))
print('----------------------------------------------------------------------')
print(groupby.sum())
print('----------------------------------------------------------------------')
print(groupby.mean())
print('----------------------------------------------------------------------')
print(df.groupby('도시').sum())
print('----------------------------------------------------------------------')

print(df.groupby('도시')['인구'].sum()) # 총인구
print(df.groupby('도시')['인구'].mean()) # 평균인구
print('----------------------------------------------------------------------')
```
