```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# 미국 질병통제센터에서 수집한
# 연도 별 신생아 성별 출생 수 데이터
# births.csv 를 이용해서 그래프 작성

births = pd.read_csv('births.csv')


# 연도별 출생수
birthyear = births.pivot_table('births',
                              'year', aggfunc='sum')

#print(birthyear)

birthyear.plot.bar()
plt.title('Year by Births')
plt.show()

# 월별 출생수
birthmon = birthyear = births.pivot_table('births',
                              'month', aggfunc='sum')
birthmon.plot.bar()
plt.show()

#print(births)

# 성별 출생수
birthgen = birthyear = births.pivot_table('births',
                              'gender', aggfunc='sum')
birthgen.plot.bar()
plt.show()

# 연도별 성별 출생수
birthyeargen = birthyear = births.pivot_table('births', 'year',
                              'gender', aggfunc='sum')
birthyeargen.plot.bar()
plt.show()

# 월별 성별 출생수
birthmongen = birthyear = births.pivot_table('births', 'month',
                              'gender', aggfunc='sum')
birthmongen.plot.bar()
plt.show()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

# 파이썬 라이브러리로 titanic 데이터에 대한 기술 통계 예제

# titanic = pd.read_csv('titanic.csv')
titanic = sns.load_dataset('titanic') # 인터넷 필요

#print(titanic['survived'])

# 승객 생존여부를 파악해서 yes/no 값을 가지는 live열 추가
titanic2['Live'] = titanic2['Survived'].apply(lambda x: 'yes' if x == 1 else 'no')
print(titanic2.info())
print(titanic2.head())
```

```
# 승객 성별 여부 파악해서 man/woman 값 가지는 Who열 추가

titanic2['Who'] = titanic2['Sex'].apply(lambda x: 'man' if x == 'male' else 'woman')
print(titanic2.info())
print('--------------------------------------------------')
print(titanic2['Who'].head())
```

```
# 승객 좌석등급여부 파악해서 First/Second/Third 값을 가지는 Seat열 추가


# apply(lambda x: 'First' if x == 1 else ('Second' if x == 2 else ('Third' if x == 3)))
# 위 lambda 식으로 구현할 수도 있겠지만, 아래 함수 def 함

def get_seats(x):
  seat = ''
  if x == '1' : seat = 'First'
  elif x == '2' : seat = 'Second'
  else : seat = 'Third'
  return seat

titanic2['Seat'] = titanic2['Pclass'].apply(lambda x : get_seats(x))


print(titanic2.info())
print('--------------------------------------------------')
print(titanic2['Seat'].head())
```

```
# 승객 나이여부 파악해서 유아/아동/10대/청년/성인 값을 가지는 Ages 열 추가
# 5 12 18 25 ~ (이하)

titanic2 = pd.read_csv('titanic.csv')

def get_Ages(x):
  Ages = ''
  if x <= 1000 : Ages = '성인'
  if x <= 25 : Ages = '청년'
  if x <= 18 : Ages = '10대'
  if x <= 12 : Ages = '아동'
  if x <= 5 : Ages = '유아'
  return Ages

def get_Ages2(x):
  Ages = ''
  if x > 25 : Ages = '성인'
  elif x > 18 : Ages = '청년'
  elif x > 12 : Ages = '10대'
  elif x > 5 : Ages = '아동'
  else : '유아'
  return Ages

def get_Ages3(x):
  Ages = ''
  if x <= 5 : Ages = '유아'
  elif x <= 12 : Ages = '아동'
  elif x <= 18 : Ages = '10대'
  elif x <= 25 : Ages = '청년'
  else : Ages = '성인'
  return Ages


titanic2['Ages'] = titanic2['Age'].apply(lambda x : get_Ages(x))
titanic2['Ages2'] = titanic2['Age'].apply(lambda x : get_Ages2(x))
titanic2['Ages3'] = titanic2['Age'].apply(lambda x : get_Ages3(x))

print(titanic2.info())
print('--------------------------------------------------')
print(titanic2['Ages'])
print('--------------------------------------------------')
print(titanic2['Ages2'])
print('--------------------------------------------------')
print(titanic2['Ages3'])
```
