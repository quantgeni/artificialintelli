```
# ex) 공부한 시간 대비 성적 예측을 단순선형회귀로 풀어봄

x = [2,4,6,8]
y = [81,93,91,97]

plt.plot(x,y,'ro')
plt.grid(True)
plt.show()

# 회귀계수를 구하기 위해 다양한 방법을 이용할 수 있는데, 여기서는 평균제곱오차MSE를 이용함

x_mean = np.mean(x)
y_mean = np.mean(y)

divisor = sum([(x-x_mean)^2 for i in x])
print(divisor)
```
