```
import numpy as np

# 배열 변환하기
# 전치, shape변환, 요소추가, 결합, 분리 등 지원

# 배열 전치transpose
# 배열객체.T
# 배열의 axis가 바뀜

a = np.arange(9).reshape(3,3)
print(a)
print(a.T)
print()

b = np.arange(24).reshape(4,6)
print(b)
print(b.T)
print()
```

```
# 배열의 shape변환
# ravel : 배열을 1차원 형태로 변경. 결과가 원본 대비 뷰이므로 특정 위치 값 수정 시 원본의 해당되는 값도 같이 변경
# reshape : 배열의 shape을 지정한 형태로 변경
```

```
c = np.arange(100, 91, -1).reshape(3,3)
d = c.ravel()
print(c)
print(d)
print(c)
```

```
d[::2] = 0
print(d)
print(c) #원본인 2차원 배열의 내용도 변경됨
```

```
e = np.arange(24).reshape(4,6)
print(e)
print()

f = e.reshape(3,4,2) #2차원 배열을 3차원으로
print(f)
```

```
# 배열크기 변경
# np.resize(대상, 크기)
# 배열의 요소 수 에 변화? resize는 reshape과 형상을 바꿀 수 있다는 측면에서는 동일하지만 요소 수에 변화를 줄 수 있는 것은 resize임

import numpy as np
import matplotlib.pyplot as plt

# 요소 수가 동일한 resize
g = np.random.randint(1,10, (2,6))
print(g)

g.resize((6,2)) # 2x6 => 6x2 배열로 수정
print(g)

# 요소 수가 변하는 resize (늘어남)
h = np.random.randint(1,50, (2,6))
print(h)

h.resize(4,6) # 2x6 => 4x6 배열로 수정
print(h) # 늘어난 요소는 0으로 채워짐

# 요소 수가 변화는 resize (줄어듬)
i = np.random.randint(1,5, (4,6))
print(i)

i.resize(2,6) # 4x6 => 2x6 배열로 수정
print(i) #줄어든 요소는 제외

# 위에 한 거 resize가 아닌, reshape으로 해보기 (아래 오류 뜸. 요소 수 줄이거나 늘이면서 형상 바꾸려면 resize()가 정답임)
i = np.random.randint(1,5, (4,6))
print(i)
i.reshape(2,6) # 오류 발생!!
print(i)
```

```
# 배열에 요소 추가
# np.append(대상, 내용, 추가방향) <- 단, concatenate()과 다르게 대상 2개만 붙일 수 있음
# 배열의 끝에 내용이 추가!

j = np.arange(1,10).reshape(3,3)
k = np.arange(9,0,-1).reshape(3,3)
t = np.arange(1,4).reshape(1,3)
y = np.arange(1,10).reshape(3,3)

print(j)
print(k)
print(t)
print()

# axis를 지정하지 않았을 때
l = np.append(j, k)
print(l) # 1차원 배열로 변환되어 추가됨
print()

l = np.append(j, np.append(k,y)) # 이렇게 하면 결국 3개 대상 결합 효과이긴 하겠으나
print(l) # 1차원 배열로 변환되어 추가됨
print()

l = np.append(j, t)
print(l) # 1차원 배열로 변환되어 추가됨
print()

# axis = 0 으로 설정 시 : 행 단위 추가
l = np.append(j, k, axis=0)
print(l) # 각각 3x3 형태를 유지하며 결합하되, 열의 shape 유지하며 결합 => 기 3x3에서 6x3이 됨
print('shape => ', l.shape)
print()

l = np.append(j, t, axis=0) #이렇게 서로 형태가 다른 것을 append 시 broadcast 안됨
print(l) # 각각 3x3 형태를 유지하며 결합하되, 열의 shape 유지하며 결합 => 기 3x3에서 6x3이 됨
print('shape => ', l.shape)
print()


# axis = 1 로 설정 시 : 열 단위 추가
l = np.append(j, k, axis=1)
print(l) # 각각 3x3 형태를 유지하며 결합하되, 행의 shape 유지하며 결합 => 기 3x3에서 3x6이 됨
print('shape => ', l.shape)
```

```
# 배열에 요소 추가
# np.insert(대상, 위치, 내용, 추가방향)
# 배열의 특정위치에 내용이 추가됨

m = np.random.randint(1,46, (3,3))
print(m)
print()

# axis를 지정하지 않았을 때
n = np.insert(m, 3, 999)
print(n) # 1차원 배열로 변환되어 추가됨
print()

# axis = 0 으로 설정 시 : 행 단위 추가
n = np.insert(m, 2, 999, axis=0)
print(n) # 추가되는 요소로 인해 나머지는 밀림 => 기 3x3에서 4x3이 됨
print('shape => ', n.shape)
print()

# axis = 1 로 설정 시 : 열 단위 추가
n = np.insert(m, 2, 999, axis=1)
print(n) # 추가되는 요소로 인해 나머지는 밀림 => 기 3x3에서 3x4가 됨
print('shape => ', n.shape)
```

```
# 배열의 요소 삭제
# np.delete(대상, 위치, 삭제방향)
# 배열의 특정위치 내용이 삭제

m = np.random.randint(1,46, (3,3))
print(m)
print()

# axis를 지정하지 않았을 때
n = np.delete(m, 2)
print(n) # 1차원 배열로 변환되어 삭제됨
print()

# axis = 0 으로 설정 시 : 행 단위 삭제
n = np.delete(m, 2, axis=0)
print(n) # 삭제되는 요소로 인해 나머지 줄어듬 => 기 3x3에서 2x3이 됨
print('shape => ', n.shape)
print()

# axis = 1 로 설정 시 : 열 단위 삭제
n = np.delete(m, 2, axis=1)
print(n) # 삭제되는 요소로 인해 나머지는 줄어듬 => 기 3x3에서 3x2가 됨
print('shape => ', n.shape)
```

```
# 배열 집합
# np.concatenate(대상1, 대상2, ...), 결합방향)
o = np.arange(1,7).reshape(2,3)
p = np.arange(9,3,-1).reshape(2,3)
print(o)
print(p)
print()

# axis 생략 시 => axis = 0 : 행 단위 결합
q = np.concatenate([o,p])
print(q)

# axis = 0 으로 지정 시 위와 동일

# axis = 1로 지정
q = np.concatenate([o,p], axis=1)
print(q)

```

```
# np.concatenate의 간략화 버전
# np.hstack(대상들)
# np.vstack(대상들)

s = np.random.randint(1,45, (2,3))
t = np.random.randint(1,45, (2,3))
u = np.random.randint(1,45, (2,3))
print(s)
print(t)
print(u)
print()

l = np.hstack([s,t,u])
print(l)
print()

l = np.vstack([s,t,u])
print(l)
```

```
# 배열 나누기
# np.hsplit(대상, 갯수)
# np.vsplit(대상, 갯수)
# 배열을 나눌 시 갯수 또는 indexing/slicing을 이용할 수 있음

v = np.arange(1,25).reshape(4,6)
print(v)
print()

# 배열 수평분할 : 열 기준으로 나눔
w = np.hsplit(v,2)
print(w)
w = np.hsplit(v,3)
print(w)
print()


# 배열 수직분할 : 행 기준으로 나눔
w = np.vsplit(v,2)
print(w)
#w = np.vsplit(v,3) # 3개 각각이 요소 수가 달라 equal division이 안된다는 오류 뜸!! 즉, 균일하게 나눠야 함
#print(w)
w = np.vsplit(v,4)
print(w)
```
