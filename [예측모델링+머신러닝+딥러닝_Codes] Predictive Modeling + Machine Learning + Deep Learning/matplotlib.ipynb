```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# matplotlib
# 파이썬의 대표적인 시각화 도구
# numpy 기반으로 만들어진 다중 플랫폼 시각화 라이브러리
# 하지만 시간이 지남에 따라 인터페이스와 스타일이 고루해짐

# R의 ggplot처럼 세렴되고 새로운 도구의 출편을 기대함
# 추후 깔끔하고 현대적인 API로 무장한 seaborn 패키지 탄생


# 매직 키워드
# 주피터 노트북에서 그래프를 작성 시 %matplotlib inline 설정해 두면 show 함수 호출없이 그래프를 그릴 수 있음

# matplotlib 인터랙티브 모드
# 명령어를 파일 단위로 한꺼번에 실행해서 결과를 보는 것이 아니라, 한줄이나 여러줄(블록) 단위로 입력, 실행하고 결과를 보는 것을 의미
# 파이썬 shell이나 주피터 노트북에서 실행하는 것을 의미
```

```
# 선그래프 : plot
# 지정한 자료는 자동적으로 y축으로 설정됨
# x축 값이 없으면 자동으로 0, 1, 2, 3, 4, 5, ...로 설정됨

data1 = pd.Series([1,4,9,16,25])
print(data1)
plt.plot(data1)
plt.show()
```

```
# x축 값 지정
np.random.seed(1906241215)
y = np.random.standard_normal(20)
x = np.arange(1,21)
print(y)
print(x)

plt.plot(y) # x축 자동 감지. 직접 넣은 argument는 y축에 해당함.
plt.show()

plt.plot(x, y) # x축 자동 감지
plt.show()

plt.plot(y, x) # x축 자동 감지
plt.show()
```

```
# 암시적 스타일 지정 : 색
# 색 : r, g, b, c, m, y, k, w

x = [1,2,3,4,5]
y = [9,8,7,6,5]

plt.plot(x,y)
plt.show()
plt.plot(x,y,'k')
plt.show()
plt.plot(x,y,'g')
plt.show()

# 암시적 스타일 지정 : 마커marker
# 데이터 위치를 나타내는 기호
# . / o / v / ^ / 1 / p / * / D
plt.plot(x, y, 'ro')
plt.show()

plt.plot(x, y, 'bD')
plt.show()

plt.plot(x, y, 'cD')
plt.show()
```

```

# 암시적 스타일 지정 : 선 모양
# 선 모양에는 실선(solid), 대시선(dashed), 점선(dotted), 대시-점선(dash-dot) 등이 있음
# : / - / -. / --

# 암시적 스타일은 최대 3개까지 표현 가능? 흠. 색/위치/모양

x = [1,2,3,4,5]
y = [9,8,7,6,5]

plt.plot(x, y, 'cD:')
plt.show()

plt.plot(x, y, 'cD-')
plt.show()

plt.plot(x, y, 'cD-.')
plt.show()

plt.plot(x, y, 'cD--')
plt.show()
```

```
# 명시적 스타일 지정
# color, linewidth, linestyle, markersize, marker edge color, marker edge width, marker face color

x = [1,2,3,4,5]
y = [9,8,7,6,5]

plt.plot(x, y, c='m', lw=3, ls='--')
plt.show()

plt.plot(x, y, c='r', lw=1, ls=':',
        marker='d', ms='20', mec='k',
        mew=5, mfc='r')
plt.show()

plt.plot(x, y, c='r', lw=1, ls=':',
        marker='d', ms='10', mec='k',
        mew=5, mfc='r')
plt.show()
```

```
# 축 지정하기
# xlim(최소, 최대), ylim(최소, 최대)을 설정

x = [1,2,3,4,5]
y = [9,8,7,6,5]

plt.plot(x, y)
plt.xlim(0,7)
plt.ylim(0,10)
plt.show()
```

```
# 축 눈금 지정하기
# 축 상의 위치 표시지점을 tick이라 함

# xticks(수치리스트, 레이블)
# yticks(수치리스트, 레이블)

x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)
#print(x)
#print()
#print(y)
plt.plot(x,y) # -1 ~ 1
plt.show()
# x축 양끝 눈금레이블이 자세히 보이지 않음
# y축 눈금레이블이 너무 자세히 표시됨

plt.plot(x,y)
plt.xticks([-3.14, -3.14/2, 0, 3.14/2, 3.14],
          ['a','b','c','d','e'])
plt.yticks([-1, -0.5, 0, 0.5, 1],
          ['A','B','C','D','E']) # 한글은 깨짐 현상 있을 수 있음


plt.grid(True)
plt.show() # 위 grid(True)는 꼭 plt.show() 이전에 해줘야!

#plt.xlim()
#plt.ylim()

# gridline : 안내선
# 틱의 위치를 잘 보여주기 위해 사용

# grid(True/False)
```

```
# 범례 추가하기 : legend
# 여러 개의 plot을 동시에 그리는 경우에 각 plot이 어떤 자료인지 나타내기 위해 사용

x = np.linspace(-3.14, 3.14, 256)
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x, y1, 'r--', label='cosine')
plt.plot(x, y2, 'b--', label='sine')
plt.grid(True)
plt.legend(loc=0) # 위 label을 미리 써줘야 함에 유의 (label = 'cosine'과 label = 'sine')
# 즉, 자동으로 범례를 표시하려면 각 그래프에 label을 지정해둬야 함

plt.show()

```
