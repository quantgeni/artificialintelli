```
# scikit-learn에서 제공하는 데이터셋
# 데이터를 불러오는 명령은 크게 3가지
# load : 내장된 데이터셋을 불러올 수 있음
# fetch : 인터넷을 통해 내려받는 대량의 데이터셋
# make : 확률분포를 이용해서 생성하는 가상 데이터

iris = load_iris()
boston = load_boston()
digits = load_digits()
```

```
# 데이터셋 속성 파악하기

print(iris.DESCR)
print(boston.DESCR)
print(digits.DESCR)
```

```
# 데이터셋 데이터 살펴보기

# print(iris) # 이건 비추. 보기 힘듬
# print(iris.data) # 데이터 (독립변수)
print(iris.target) # 레이블 (종속변수)
print(iris.target_names) # 레이블 범주값
print(iris.feature_names) # 독립변수명
```

```
df = pd.DataFrame(iris.data, columns=iris.feature_names)
# df = pd.DataFrame(iris.data)
print(df.head())
print('----------------------------------')
sr = pd.Series(iris.target, dtype='category')
print(sr.head()) # 당초 iris.target이 dict에서 0, 1, 2의 value를 가졌으므로 그런 것들로 이루어진게 나옴
print('----------------------------------')
sr = sr.cat.rename_categories(iris.target_names) # 0, 1, 2에 상응한 target_names의 value로 바꿈 (즉, 'setosa' 'versicolor' 'virginica'로 매칭)
print(sr.head())
print('----------------------------------')
df['species'] = sr
print(df.head())
```

```
scatter_matrix(df) # 산점도 행렬
plt.show()
```

```
sns.pairplot(df) # 산점도 행렬
plt.show()
```

```
sns.pairplot(df, hue='species') # 산점도 행렬. 좀 예쁘게 나옴. 씨본 짱
plt.show()
```

```
# 가상 데이터 생성하기
# 분류 모형의 테스트를 위해
# make_blob, make_classification, make_moon, make_circle 등의
# 가상 데이터 생성 함수를 지원함

from sklearn.datasets import make_classification
#from sklearn.datasets import make_blob
#from sklearn.datasets import make_moon
#from sklearn.datasets import make_circle

print('--------------------------------------------------------------------------------------- 독립변수 1개')

X, y = make_classification(n_features=1, # 독립변수 수. 기본은 20
                           n_informative=1, # 독립변수와 종속변수 간 상관관계 수
                           n_redundant=0, # 독립변수와 종속변수 간 선형관계 수
                           n_clusters_per_class=1, #군집수
                           random_state=1906261730) #난수생성초기값
# n_samples : 샘플수, 기본값 100

# => 1개의 독립변수를 가지는 가상데이터

plt.grid(True)
plt.scatter(X, y, marker='o', c=y, # y에 맞춰 색깔 나오게 / edgecolors는 빼도 상관 없지만 있어야 각 marker가 잘보임 ㅎㅎ 
           s=100, edgecolors='k')

plt.show()
```

```
from sklearn.datasets import make_classification
#from sklearn.datasets import make_blob
#from sklearn.datasets import make_moon
#from sklearn.datasets import make_circle

print('--------------------------------------------------------------------------------------- 독립변수 2개')

X, y = make_classification(n_features=2, # 독립변수 수. 기본은 20
                           n_informative=1, # 독립변수와 종속변수 간 상관관계 수
                           n_redundant=0, # 독립변수와 종속변수 간 선형관계 수
                           n_clusters_per_class=1,
                           random_state=1906261730)

# => 1개의 독립변수를 가지는 가상데이터

#print(X)
#print(X[:,0])

plt.grid(True)
plt.scatter(X[:,0], X[:,1], marker='o', c=y, # y에 맞춰 색깔 나오게 / edgecolors는 빼도 상관 없지만 있어야 각 marker가 잘보임 ㅎㅎ <- 독립변수 2개 (n_features=2)일 때는 변화줘야
           s=100, edgecolors='k') # y는 빼줌

plt.show()
```

```
# make_blobs
# 가우시안 정규분포를 이용해서 군집용 가상 데이터 생성

from sklearn.datasets import make_blobs

X, y = make_blobs(n_features=2, centers=3, random_state=1906261750) # centers = 생성할 군집수 (기본은 3)
plt.scatter(X[:,0], X[:,1], marker='o', c=y,
           s=100, edgecolors='k')

plt.grid(True)
plt.show()
```
